```python
import os
import cv2
import numpy as np
from scipy import ndimage as ndi
from skimage import feature
from matplotlib import pyplot as plt
from PIL import Image as pil_image

from google.colab.patches import cv2_imshow
from google.colab import drive, files
```

```python
def read_images(path_to_dir):
    directory = sorted(os.listdir(path_to_dir))
    return [pil_image.open(path_to_dir + '/' + file_name) for file_name in directory]
```

# Separate text from background

```python
images = read_images(main_dir + 'text')
```

```python
images[3]
```

```python
def extract_text_img_3(img):
    grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    threshold = cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 5)
    image = cv2.fastNlMeansDenoising(threshold,None,80,5,15)
    cv2_imshow(image)
    
extract_text_img_3(np.array(images[3]))
```

```python
def extract_text_img_4(img):
    grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    grayscaled[grayscaled < 70] = 0
    cv2_imshow(255 - grayscaled)
    
extract_text_img_4(np.array(images[4].transpose(pil_image.ROTATE_270)))
```
